```python
import pandas as pd
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
column_name = ["SOAP_S", "SOAP_O", "SOAP_A", "SOAP_P"]
```

```python
file_name = "data\\201709_1231.xls"
```

```python
xl = pd.ExcelFile(file_name )
```

```python
sheet = xl.sheet_names[0]
```

```python
df = xl.parse(header=0,index_col=None,parse_cols="A,B,C,D")
```

```python
df.head
```

```python
df.shape
```

```python
df.dropna()
```

```python
df_drop_na = df.dropna()
```

```python
S = df_drop_na["SOAP_S"].str.replace("【Ｓ】","").str.replace('\r','').str.replace('\n','')
```

```python
O = df_drop_na["SOAP_O"].str.replace("【Ｏ】","").str.replace('\r','').str.replace('\n','')
```

```python
A = df_drop_na["SOAP_A"].str.replace("【Ａ】","").str.replace('\r','').str.replace('\n','')
```

```python
P = df_drop_na["SOAP_P"].str.replace("【Ｐ】","").str.replace('\r','').str.replace('\n','')
```

```python
clean_data = pd.concat([S,O,A,P], axis=1)
```

```python
clean_data
```

```python
clean_data.to_csv("data\\csv\\201709_1231.csv",encoding='utf-8',index=False)
```

```python
data_size = clean_data.shape[0]
chinese_indices = set()
for i in clean_data.index.tolist():
    for column in column_name:
        if not isEnglish(clean_data[column][i]):
            chinese_indices.add(i)
```

```python
english_data = clean_data.drop(chinese_indices)
```

```python
english_data
```

```python
english_data.to_csv("data\\csv\\201709_1231_eng.csv",encoding='utf-8',index=False)
```
