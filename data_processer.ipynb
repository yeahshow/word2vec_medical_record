```python
import pandas as pd
import uuid
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
def uuid_generator(s):
    return str(uuid.uuid3(uuid.NAMESPACE_DNS, s))
```

```python
column_name = ["SOAP_S", "SOAP_O", "SOAP_A", "SOAP_P"]
```

```python
file_name = "data\\201709_1231.xls"
```

```python
xl = pd.ExcelFile(file_name )
```

```python
sheet = xl.sheet_names[0]
```

```python
df = xl.parse(header=0,index_col=None,parse_cols="A,B,C,D,F,H,J,K")
```

```python
df.shape
```

```python
df['unique_string'] = df[df.columns[4:]].apply(lambda x: '-'.join(x.astype(str)),axis=1)
```

```python
df['uuid'] = df['unique_string'].apply(lambda x: uuid_generator(x))
```

```python
df
```

```python
df_drop_na = df.dropna()
```

```python
S = df_drop_na["SOAP_S"].str.replace("【Ｓ】","").str.replace('\r','').str.replace('\n','')
```

```python
O = df_drop_na["SOAP_O"].str.replace("【Ｏ】","").str.replace('\r','').str.replace('\n','')
```

```python
A = df_drop_na["SOAP_A"].str.replace("【Ａ】","").str.replace('\r','').str.replace('\n','')
```

```python
P = df_drop_na["SOAP_P"].str.replace("【Ｐ】","").str.replace('\r','').str.replace('\n','')
```

```python
clean_data = pd.concat([S,O,A,P, df_drop_na["uuid"]], axis=1)
```

```python
clean_data
```

```python
writer = pd.ExcelWriter('data\\\processed\\201709_1231_p.xls')
```

```python
# clean_data.to_csv("data\\\processed\\201709_1231.csv",encoding='utf-8',index=False)
clean_data.to_excel(writer,'non_eng')
```

```python
data_size = clean_data.shape[0]
chinese_indices = set()
for i in clean_data.index.tolist():
    for column in column_name:
        if not isEnglish(clean_data[column][i]):
            chinese_indices.add(i)
```

```python
english_data = clean_data.drop(chinese_indices)
```
