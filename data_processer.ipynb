```python
import pandas as pd
import matplotlib.pyplot as plt
import uuid
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
def uuid_generator(s):
    return str(uuid.uuid3(uuid.NAMESPACE_DNS, s))
```

```python
#column_name = ["SOAP_S", "SOAP_O", "SOAP_A", "SOAP_P"]
file_name = "data\\201709_1231.xls"
#xl = pd.ExcelFile(file_name )
#sheet = xl.sheet_names[0]
```

```python
df = pd.read_excel(file_name, sheet_name=0, header=0, index_col=None, usecols = 'A:H')
#df = xl.parse(header=0,index_col=None,parse_cols="A,B,C,D")
```

```python
df.shape
```

```python
df['uuid'] = df[df.columns[5:7]].apply(lambda x: '-'.join(x.astype(str)),axis=1).apply(lambda x: uuid_generator(x))
```

```python
df.head(10)
```

```python
df_drop_na = df.dropna()

S = df_drop_na["SOAP_S"].str.replace('\r','').str.replace('\n','')

O = df_drop_na["SOAP_O"].str.replace('\r','').str.replace('\n','')

A = df_drop_na["SOAP_A"].str.replace('\r','').str.replace('\n','')

P = df_drop_na["SOAP_P"].str.replace('\r','').str.replace('\n','')

PROTOCOL = df_drop_na["PROTOCOL"].str.replace('\r','').str.replace('\n','')

clean_data = pd.concat([S,O,A,P,PROTOCOL, df_drop_na["uuid"]], axis=1)

clean_data
```

```python
clean_data_no_chinese = pd.DataFrame()
for col in clean_data.columns:
    clean_data_no_chinese[col]= clean_data[col].str.replace(r'[^\x00-\x7F]+', '')

clean_data_no_chinese
```

```python
writer = pd.ExcelWriter(file_name + 'proccessed.xls')

clean_data.to_excel(writer,'clean')
clean_data_no_chinese.to_excel(writer,'eng_only')

writer.save()
```
