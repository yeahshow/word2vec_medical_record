```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import uuid
%matplotlib inline
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
def uuid_generator(s):
    return str(uuid.uuid3(uuid.NAMESPACE_DNS, s))
```

```python
df1 = pd.read_excel('20170101_430.xls', sheet_name=0, header=0, index_col=None)
df2 = pd.read_excel('20170431_831.xls', sheet_name=0, header=0, index_col=None)
df3 = pd.read_excel('201709_1231.xls', sheet_name=0, header=0, index_col=None)
df4 = pd.read_excel('20180101_0430.xls', sheet_name=0, header=0, index_col=None)
df = pd.concat([df1,df2,df3,df4])
```

```python
df.shape
```

```python
df.head(10)
```

```python
df['uuid'] = df[df.columns[[5, 7, 9]]].apply(lambda x: '-'.join(x.astype(str)),axis=1).apply(lambda x: uuid_generator(x))
```

```python
df.head(10)
```

```python
df_drop_na = df.drop('排程日期',axis='columns').dropna()

S = df_drop_na["SOAP_S"].str.replace('\r','').str.replace('\n','')

O = df_drop_na["SOAP_O"].str.replace('\r','').str.replace('\n','')

A = df_drop_na["SOAP_A"].str.replace('\r','').str.replace('\n','')

P = df_drop_na["SOAP_P"].str.replace('\r','').str.replace('\n','')

PROTOCOL = df_drop_na["PROTOCOL"].str.replace('\r','').str.replace('\n','')

clean_data = pd.concat([S,O,A,P,PROTOCOL, df_drop_na["uuid"]], axis=1)

clean_data
```

```python
clean_data_no_chinese = pd.DataFrame()
for col in clean_data.columns:
    clean_data_no_chinese[col]= clean_data[col].str.replace(r'[^\x00-\x7F]+', '')

clean_data_no_chinese
```

```python
writer = pd.ExcelWriter('proccessed.xls')

clean_data.to_excel(writer,'clean')
clean_data_no_chinese.to_excel(writer,'eng_only')

writer.save()
```
